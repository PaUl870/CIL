```python
data_manager.load_and_split()


def calculate_exclusivity_stats(train_df, valid_df, wishlist_df):
    # Sets of IDs from Wishlist
    wish_sids = set(wishlist_df["sid"].unique())
    wish_pids = set(wishlist_df["pid"].unique())

    # Sets of IDs from Ratings
    train_sids = set(train_df["sid"].unique())
    train_pids = set(train_df["pid"].unique())
    
    val_sids = set(valid_df["sid"].unique())
    val_pids = set(valid_df["pid"].unique())

    # 1. Wishlist-only relative to Training
    # (These entities have NO purchase signals for the model to learn from)
    wish_only_train_users = wish_sids - train_sids
    wish_only_train_items = wish_pids - train_pids

    # 2. Wishlist-only relative to Validation
    # (These entities are in the wishlist but have no target rating in the val set)
    wish_only_val_users = wish_sids - val_sids
    wish_only_val_items = wish_pids - val_pids

    # 3. Global Exclusivity (Never rated in either set)
    all_rated_sids = train_sids.union(val_sids)
    all_rated_pids = train_pids.union(val_pids)
    
    global_wish_only_users = wish_sids - all_rated_sids
    global_wish_only_items = wish_pids - all_rated_pids

    print("--- Wishlist Exclusivity Report ---")
    print(f"Total Unique Users in Wishlist: {len(wish_sids)}")
    print(f"Total Unique Items in Wishlist: {len(wish_pids)}")
    print("-" * 35)
    print(f"Users with ONLY wishlist (not in Train): {len(wish_only_train_users)}")
    print(f"Items with ONLY wishlist (not in Train): {len(wish_only_train_items)}")
    print("-" * 35)
    print(f"Users with ONLY wishlist (not in Val):   {len(wish_only_val_users)}")
    print(f"Items with ONLY wishlist (not in Val):   {len(wish_only_val_items)}")
    print("-" * 35)
    print(f"GLOBAL: Users with NO ratings anywhere:  {len(global_wish_only_users)}")
    print(f"GLOBAL: Items with NO ratings anywhere:  {len(global_wish_only_items)}")    
    print("-" * 35)
    print(f"GLOBAL: Users with NO ratings anywhere:  {len(wish_only_train_users - wish_only_val_users)}")
    print(f"GLOBAL: Items with NO ratings anywhere:  {len(wish_only_train_items - wish_only_val_items)}")

# Usage:
calculate_exclusivity_stats(data_manager.train_df, data_manager.valid_df, data_manager.wishlist_df)
```

```python
def val_warm_fn(model):
    model.eval()
    with torch.no_grad():
        val_preds = model.predict_ratings(
            val_warm["sid"].values, 
            val_warm["pid"].values,
            edge_t0_index,    # The graph structure
            edge_t1_index,
            edge_t0_weights   # The graph weights
        )
        val_preds = np.clip(val_preds, 1.0, 5.0)
        
        return root_mean_squared_error(val_warm["rating"].values, val_preds)


def val_cold_fn(model):
    model.eval()
    with torch.no_grad():
        val_preds = model.predict_ratings(
            val_cold["sid"].values, 
            val_cold["pid"].values,
            edge_t0_index,    # The graph structure
            edge_t1_index,
            edge_t0_weights   # The graph weights
        )
        val_preds = np.clip(val_preds, 1.0, 5.0)
        
        return root_mean_squared_error(val_cold["rating"].values, val_preds)



def val_fn(model):
    model.eval()
    with torch.no_grad():
        val_preds = model.predict_ratings(
            data_manager.valid_df["sid"].values, 
            data_manager.valid_df["pid"].values,
            edge_t0_index,    # The graph structure
            edge_t1_index,
            edge_t0_weights   # The graph weights
        )
        val_preds = np.clip(val_preds, 1.0, 5.0)
        
        return root_mean_squared_error(data_manager.valid_df["rating"].values, val_preds)
```

```python
import torch
from collections import defaultdict


def _get_emb(model):
    if hasattr(model, "user_embedding"):                        # GCN
        return (model.user_embedding.weight.data,
                model.item_embedding.weight.data,
                model.num_users,
                model.num_items)
    else:                                                       # ALS
        return (model.user_emb.data,
                model.item_emb.data,
                model.user_emb.shape[0],
                model.item_emb.shape[0])


@torch.no_grad()
def initialize_cold_start(model, wishlist_df, cold_users, cold_items, top_k: int = 5):
    user_emb, item_emb, num_users, num_items = _get_emb(model)
    device = user_emb.device

    user_wishlist = defaultdict(set)
    item_wishlist = defaultdict(set)
    for sid, pid in zip(wishlist_df["sid"].values, wishlist_df["pid"].values):
        user_wishlist[int(sid)].add(int(pid))
        item_wishlist[int(pid)].add(int(sid))

    def jaccard_aggregate(cold_entities, entity_wishlist,
                          counterpart_wishlist, warm_mask_set, emb_matrix):
        updated, skipped = 0, 0
        for eid in cold_entities:
            cold_set = entity_wishlist.get(eid, set())
            if not cold_set:
                skipped += 1
                continue

            overlap_counts = defaultdict(int)
            for counterpart in cold_set:
                for warm_eid in counterpart_wishlist.get(counterpart, set()):
                    if warm_eid in warm_mask_set:
                        overlap_counts[warm_eid] += 1

            if not overlap_counts:
                skipped += 1
                continue

            cold_size    = len(cold_set)
            similarities = {
                v: ov / (cold_size + len(entity_wishlist[v]) - ov)
                for v, ov in overlap_counts.items()
            }
            top = sorted(similarities, key=similarities.__getitem__, reverse=True)[:top_k]
            w   = torch.tensor([similarities[v] for v in top],
                               dtype=torch.float, device=device)
            w   = w / w.sum()
            idx = torch.tensor(top, dtype=torch.long, device=device)
            emb_matrix[eid] = (w.unsqueeze(1) * emb_matrix[idx]).sum(0)
            updated += 1

        return updated, skipped

    warm_users = set(range(num_users)) - cold_users
    warm_items = set(range(num_items)) - cold_items

    updated_u, skipped_u = jaccard_aggregate(
        list(cold_users), user_wishlist, item_wishlist, warm_users, user_emb
    )
    updated_i, skipped_i = jaccard_aggregate(
        list(cold_items), item_wishlist, user_wishlist, warm_items, item_emb
    )

    print(f"Users: {updated_u} initialized, {skipped_u} skipped")
    print(f"Items: {updated_i} initialized, {skipped_i} skipped")
```

```python
import torch
import numpy as np
from collections import defaultdict
from sklearn.cluster import KMeans


def _get_emb(model):
    """Returns (user_emb_data, item_emb_data, num_users, num_items)"""
    if hasattr(model, "user_embedding"):                        # GCN
        return (model.user_embedding.weight.data,
                model.item_embedding.weight.data,
                model.num_users,
                model.num_items)
    else:                                                       # ALS
        return (model.user_emb.data,
                model.item_emb.data,
                model.user_emb.shape[0],
                model.item_emb.shape[0])


@torch.no_grad()
def initialize_cold_start(model, wishlist_df, cold_users, cold_items,
                           n_clusters: int = 50):
    user_emb, item_emb, num_users, num_items = _get_emb(model)
    device = user_emb.device

    user_wishlist = defaultdict(set)
    item_wishlist = defaultdict(set)
    for sid, pid in zip(wishlist_df["sid"].values, wishlist_df["pid"].values):
        user_wishlist[int(sid)].add(int(pid))
        item_wishlist[int(pid)].add(int(sid))

    def cluster_aggregate(cold_entities, entity_wishlist, counterpart_wishlist,
                          warm_mask_set, emb_matrix):
        warm_list = sorted(warm_mask_set)
        warm_embs = emb_matrix[warm_list].cpu().numpy()

        k = min(n_clusters, len(warm_list))
        km = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = km.fit_predict(warm_embs)
        centroids = torch.tensor(km.cluster_centers_,
                                 dtype=torch.float, device=device)

        cluster_item_counts = defaultdict(lambda: defaultdict(int))
        for local_idx, warm_eid in enumerate(warm_list):
            cluster_id = labels[local_idx]
            for item in entity_wishlist.get(warm_eid, set()):
                cluster_item_counts[cluster_id][item] += 1

        updated, skipped = 0, 0
        for eid in cold_entities:
            cold_set = entity_wishlist.get(eid, set())
            if not cold_set:
                skipped += 1
                continue

            cluster_scores = np.zeros(k)
            for cluster_id in range(k):
                for item in cold_set:
                    cluster_scores[cluster_id] += cluster_item_counts[cluster_id].get(item, 0)

            if cluster_scores.max() == 0:
                skipped += 1
                continue

            scores_t = torch.tensor(cluster_scores, dtype=torch.float, device=device)
            weights  = scores_t / scores_t.sum()
            emb_matrix[eid] = (weights.unsqueeze(1) * centroids).sum(0)
            updated += 1

        return updated, skipped

    warm_users = set(range(num_users)) - cold_users
    warm_items = set(range(num_items)) - cold_items

    updated_u, skipped_u = cluster_aggregate(
        list(cold_users), user_wishlist, item_wishlist, warm_users, user_emb
    )
    updated_i, skipped_i = cluster_aggregate(
        list(cold_items), item_wishlist, user_wishlist, warm_items, item_emb
    )

    print(f"Users: {updated_u} initialized, {skipped_u} skipped")
    print(f"Items: {updated_i} initialized, {skipped_i} skipped")
```
